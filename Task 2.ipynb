```python
data2=data.fillna(value = 0)
data2
```

```python
data3=data.fillna(method="pad")
data3
```

```python
data4=data.fillna(method="bfill")
data4
```

```python
one_hot_encoded_data = pd.get_dummies(data4, columns = ['ethnicity', 'gender', 'Class/ASD']) 
print(one_hot_encoded_data)
```

```python
sb.histplot(one_hot_encoded_data['gender_m'])
```

```python
df=data4[data4['Class/ASD'].str.contains("YES")]
print(df)
```

```python
df.head(30)
```

```python
df['Sum_A_Scores'] = df[['A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score', 'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'A10_Score']].sum(axis=1)
print(df['Sum_A_Scores'])
      
```

```python
filtered_df = df[df['Sum_A_Scores'] < 7]
print(filtered_df)
```

```python
dff=data4[data4['Class/ASD'].str.contains("NO")]
dff.head(30)
```

```python
dff['Sum_A_Scores'] = dff[['A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score', 'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'A10_Score']].sum(axis=1)
print(dff['Sum_A_Scores'])
```

```python
filtered_dff = dff[dff['Sum_A_Scores'] >6]
print(filtered_dff)
```

<h1>Pattern</h1>
Those patients with A_Score >=7 is screened with autism, and those with less than 7 is autism free regardless or any other classifications.
